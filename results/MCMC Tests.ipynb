```julia
using Random
using Plots
using Distributions
using AdaptiveSurveillance
```

```julia
const L = 1
const ν = 1 / 21 # approx 6 months until there is an outbreak
const Γd = [Geometric(ν) for l = 1:L]
const p0 = 0.01 * ones(L)
const β = 4e-6 * 536 * 7
const p = repeat(prevalance_sequence(p0[1], β), 1, L)
const n = 200
```

```julia
α = Inf
```

```julia
print(p[1:200])
```

```julia
t, la, false_alarm, delay, test_data, locations_visited, ntimes_visisted, z, w = replication(L, Γd, [100], p0, p, n, 
    astat_isotonic, α, tpolicy_random, nothing,
    rng1 = MersenneTwister(1), rng2 = MersenneTwister(2), maxiters = 1000);
```

```julia
n_samples = 1000
initialized = zeros(Bool, L)
mcmc_samples = zeros(n_samples, 3, L)
beta_parameters = ones(L, 2)
tstate = tstate_evsi(n_samples, initialized, mcmc_samples, beta_parameters)
```

```julia
AdaptiveSurveillance.logistic_samples!(n, tstate, MersenneTwister(1), test_data, locations_visited, t, 1)
```
