```julia
using Distributions
using Plots
using AdaptiveSurveillance
```

```julia
const output_path = joinpath(dirname(pathof(AdaptiveSurveillance)), "..", "results", "tmp")
const ν = 1 / 21 # approx 6 months until there is an outbreak
const Γd = Geometric(ν)
const p0 = 0.01
const β = 4e-6 * 536 * 7
const p = prevalance_sequence(p0[1], β)
const n = 200
```

```julia
α = 1000 # the higher, the less false positives
```

```julia
T = length(p)
plot((1:T)./52, p, ylim=(0, 1), legend=false, xlabel = "Year", ylabel = "Prevalence")
```

```julia
# Delay for different thresholds
println(length(p[p .<= 2 * p0]))
println(length(p[p .<= 3 * p0]))
println(length(p[p .<= 5 * p0]))
println(length(p[p .<= 0.10]))
```

```julia
plot(1:T, cdf.(Γd, 0:(T-1)))
```

```julia
# Debug
t, false_alarm, delay, test_data, z, thres = sp_single.replication(Γd, 10, p0, p, n, sp_single.apolicy_isotonic, α);
```

```julia
plot(1:(t-1), test_data[1:(t-1)], xlabel = "Time", ylabel = "Positive Count", legend=false)
```

```julia
plot(2:(t-1), hcat(z[3:t], thres[3:t]), xlabel = "Time", ylabel = "Value", label = ["Alarm Statistic" "Threshold"],
legend=:topleft)
```

```julia
α = 1000
```

```julia
# Alarm time distribution
atd_T = 200
atd_0 = sp_single.alarm_time_distribution(Int(1e4), Γd, 0, p0, p, n, sp_single.apolicy_isotonic, α, maxiters=atd_T);
atd_0 = cumsum(atd_0) ./ sum(atd_0);
```

```julia
atd_1 = sp_single.alarm_time_distribution(Int(1e4), Γd, typemax(Int64), p0, p, n, sp_single.apolicy_isotonic, α, maxiters=atd_T);
atd_1 = cumsum(atd_1) ./ sum(atd_1);
```

```julia
plot(1:atd_T, hcat(atd_0[1:atd_T], atd_1[1:atd_T]), label = ["Change at 0" "No Change"], legend=:topleft)
```

```julia
# Predictive Value
pvt = sp_single.predictive_value(Int(1e4), 21, 52, Γd, p0, p, n, sp_single.apolicy_isotonic, α);
```

```julia
plot(pvt, xlabel = "Alarm Time", ylabel = "PV", legend=false, xlim=(21, 52))
```

```julia
# Probability of success
T_end = 21
sp2 = sp_single.probability_successfull_detection(Int(1e4), T_end, 47, Γd, p0, p, n, sp_single.apolicy_isotonic, α);
sp3 = sp_single.probability_successfull_detection(Int(1e4), T_end, 75, Γd, p0, p, n, sp_single.apolicy_isotonic, α);
sp5 = sp_single.probability_successfull_detection(Int(1e4), T_end, 110, Γd, p0, p, n, sp_single.apolicy_isotonic, α);
```

```julia
plot(1:T_end, hcat(sp2, sp3), label = ["2x" "3x"], xlabel = "Outbreak Start Time", ylabel = "PSD", ylim=(0, 1.0))
```
